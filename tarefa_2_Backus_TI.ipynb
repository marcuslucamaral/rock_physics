## Parte 01


####Definindo as funções
Funções de backus retiradas de: https://pangea.stanford.edu/departments/geophysics/dropbox/SRB/public/data/RPHtools.htm

Velocidades (VP, VSH, VSV) no ângulo theta a partir do eixo de simetria de um meio TI a partir de sua rigidez elástica:
$\begin{aligned} V_p(\theta) & =\sqrt{\frac{C_{11} \sin ^2(\theta)+C_{33} \cos ^2(\theta)+C_{55}+\sqrt{M(\theta)}}{2 \rho}} \\ V_{S V}(\theta) & =\sqrt{\frac{C_{11} \sin ^2(\theta)+C_{33} \cos ^2(\theta)+C_{55}-\sqrt{M(\theta)}}{2 \rho}} \\ V_{S H} & =\sqrt{\frac{C_{66} \sin ^2(\theta)+C_{55} \cos ^2(\theta)}{\rho}} \\ M(\theta) & =\left[\left(C_{11}-C_{55}\right) \sin ^2(\theta)-\left(C_{33}-C_{55}\right) \cos ^2(\theta)\right]^2+\left(C_{13}+C_{55}\right)^2 \sin ^2(2 \theta)\end{aligned}$

```python
def Ciso(rho, vp, vs):

    # stiffness parameters for an isotropic rock
    # rho - Density
    # vp - compressional velocity
    # vs - shale velocity
    # K  - Isotropic rock bulk modulus
    # mu - Isotropic rock shear modulus

    mu = rho * vs**2
    lam = rho * vp**2 - 2 * mu

    c11 = lam + 2 * mu
    c12 = lam
    c44 = mu

    c = np.array([c11, c11, c44, c44, c12])

    return c

```

```python
def bkus(frac, vp, vs, density):
    # Verifique se a soma das frações é próxima de 1
     #se não np.isclose(np.sum(f), 1, atol=1e-4):
     # print('A soma das frações não é igual a 1!')
     # retorn none, none, none

     # Força f a somar a unidade, normalizando
    frac = frac / np.sum(frac)

    mu = density * vs**2
    lam = density * vp**2 - 2 * mu

    x = np.sum(frac * mu * (lam + mu) / (lam + 2 * mu))
    y = np.sum(frac * mu * lam / (lam + 2 * mu))
    z = np.sum(frac * lam / (lam + 2 * mu))
    u = np.sum(frac / (lam + 2 * mu))
    v = np.sum(frac / mu)
    w = np.sum(frac * mu)

    c11 = 4 * x + z**2 / u
    c12 = 2 * y + z**2 / u
    c33 = 1 / u
    c13 = z / u
    c44 = 1 / v
    c66 = w
    c661 = (c11 - c12) / 2

    if np.abs(c661 - c66) > 0.1:
        print('There is a problem with the calculation')
        print('c66 is not equal to (c11-c12)/2')
        return None, None, None

    coeff = np.array([c11, c33, c44, c66, c13])

    rho = np.sum(frac * density)


    return coeff, rho
```

```python
def bkuslog(depth, vp, vs, den):
    # Calcula a média do Backus a partir dos dados de registro, assumindo que cada ponto de profundidade
     # corresponde a uma fina camada isotrópica. A média está em todo o registro
     # fornecido como entrada.
     # Médias de volume simples de densidade e porosidade também são retornadas.

     # Calcular vetor de espessuras de camada. Suponha que cada ponto de profundidade seja o
     # topo da camada fina. Extrapole outra camada fina na parte inferior.
    thick = np.diff(depth)
    thick = np.append(thick, thick[-1])

  # Divida pela espessura total para obter frações de volume. Isso também
     # retorna os valores para números positivos, caso as profundidades tenham sido inseridas
     # Em ordem inversa.
    tot = np.sum(thick)
    thick = np.abs(thick / tot)

# chama Rotina media de backus
    c, rho = bkus(thick, vp, vs, den)

    return c, rho
```

```python
## Parâmetros Anisotrópicos
## Da função bkus, temos que cc = (c11, c33, c44, c66, c13)

def parametros(cc, rho):
    vp0     = np.sqrt(cc[1]/rho)
    vs0     = np.sqrt(cc[2]/rho)
    epsilon = (cc[0] - cc[1])/(2*cc[1])
    gama    = (cc[3] - cc[2])/(2*cc[2])
    delta   = ((cc[4] + cc[2])**(2)-(cc[1] - cc[2])**(2))/(2*cc[1]*(cc[1] - cc[2]))

    return vp0, vs0, epsilon, gama, delta
```

```python
# janelas de aproximadamente 10, 25, 50 e 100 m

def janelas(tamanho, dado):
    df = dado.copy()
    depth = df.index.to_numpy()

    limite     = (depth[-1] - depth[0])
    num_janelas = np.floor(limite/tamanho)   # Utilizei o "floor" ao invés do "round" para garantir que a operação vai ser realizada dentro do limites de cada dado.
    janela = limite/num_janelas
    num_camadas = int(np.floor(limite/janela))


    return janela, num_camadas, depth
```

```python
# Média para todo o registro (uma única camada efetiva)

def media_log(vp, vs, rho, depth, num_camadas, janela):

    vp_dado, vs_dado, rho_dado, camadas = np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas)

    i = 0
    index = 0

    for itens in range(num_camadas):
        if i < (len(depth) - int(janela)):
            vp_dado[index]  = np.mean(vp[i:(i+int(len(depth)/num_camadas))])
            vs_dado[index]  = np.mean(vs[i:(i+int(len(depth)/num_camadas))])
            rho_dado[index] = np.mean(rho[i:(i+int(len(depth)/num_camadas))])
            camadas[index]  = depth[0] + index*janela

            index = index+1
            i = i + int(len(depth)/num_camadas)

    return vp_dado, vs_dado, rho_dado, camadas
```

```python
# Médias e parâmetros anisotrópicos para cada camada isotrópica
def media_camadas_iso(vp, vs, rho, depth, num_camadas, janela):

    limite = int((depth[-1] - depth[0])/janela)

    vp_dado, vs_dado, rho_dado, camadas = np.zeros(limite), np.zeros(limite), np.zeros(limite), np.zeros(limite)
    vp0, vs0, epsilon, gama, delta, rhoc = np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas)

    i = 0
    for beds in range(num_camadas):
        for item in range(limite):
            if i < len(depth):
                vp_dado[item]  = vp[i]
                vs_dado[item]  = vs[i]
                rho_dado[item] = rho[i]
                camadas[item]  = depth[i]
                i = i + 1

        #c, rhobkus = bkuslog(camadas[1:-2], vp_dado[1:-2], vs_dado[1:-2], rho_dado[1:-2])  #para quando as camadas são anisortrópicas VTI
        c = Ciso(np.mean(rho_dado), np.mean(vp_dado), np.mean(vs_dado))
        rhobkus = np.mean(rho_dado)
        vp0[beds], vs0[beds], epsilon[beds], gama[beds], delta[beds] = parametros(c, rhobkus)
        rhoc[beds] = np.mean(rho_dado)


    return vp0, vs0, epsilon, gama, delta, rhoc  # para cada camada, se isotrópica: epsilon, gama e delta = 0
```

```python
# Médias e parâmetros anisotrópicos para cada camada ANISOTRÓPICA
def media_camadas_ani(vp, vs, rho, depth, num_camadas, janela):

    limite = int((depth[-1] - depth[0])/janela)

    vp_dado, vs_dado, rho_dado, camadas = np.zeros(limite), np.zeros(limite), np.zeros(limite), np.zeros(limite)
    vp0, vs0, epsilon, gama, delta = np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas)

    i = 0
    for beds in range(num_camadas):
        for item in range(limite):
            if i < len(depth):
                vp_dado[item]  = vp[i]
                vs_dado[item]  = vs[i]
                rho_dado[item] = rho[i]
                camadas[item]  = depth[i]
                i = i + 1

        c, rhobkus = bkuslog(camadas[1:-2], vp_dado[1:-2], vs_dado[1:-2], rho_dado[1:-2])
        vp0[beds], vs0[beds], epsilon[beds], gama[beds], delta[beds] = parametros(c, rhobkus)


    return vp0, vs0, epsilon, gama, delta
```

```python
# Resultados para cada camada

# Norne2
#entre com o tamanho da janela: 10,25, 50 ou 100
janela2, num_camadas2, depth2 = janelas(10, norne2)
VP_N2, VS_N2, RHO_N2, CAMADAS_N2 = media_log(VP2, VS2, RHOB2, depth2, num_camadas2, janela2)
VP02C, VS02C, EPSILON2C, GAMA2C, DELTA2C, RHO2C = media_camadas_iso(VP2, VS2, RHOB2, depth2, num_camadas2, janela2)

# Norne9

#entre com o tamanho da janela: 10,25, 50 ou 100
janela9, num_camadas9, depth9 = janelas(10, norne9)
VP_N9, VS_N9, RHO_N9, CAMADAS_N9 = media_log(VP9, VS9, RHOB9, depth9, num_camadas9, janela9)
VP09C, VS09C, EPSILON9C, GAMA9C, DELTA9C, RHO9C = media_camadas_iso(VP9, VS9, RHOB9, depth9, num_camadas9, janela9)
```

### Funções utilizadas

```python
def bkusc(f, vp, vs, den):
    # Força f a somar a unidade, normalizando
    f = f / np.sum(f)

    # Calcula lambda e mu isotrópicos para cada camada fina
    mu = den * vs**2
    lam = den * vp**2 - 2 * mu

    # Executar médias Backus
    x = np.sum(f * mu * (lam + mu) / (lam + 2 * mu))
    y = np.sum(f * mu * lam / (lam + 2 * mu))
    z = np.sum(f * lam / (lam + 2 * mu))
    u = np.sum(f / (lam + 2 * mu))
    v = np.sum(f / mu)
    w = np.sum(f * mu)

    # Atribuir ao formato 6x6
    c11 = 4 * x + z * z / u
    c12 = 2 * y + z * z / u
    c33 = 1 / u
    c13 = z / u
    c44 = 1 / v
    c66 = w

    cvti = np.array([c11, c33, c44, c66, c13])
    chti = np.array([c33, c11, c66, c44, c12])

    rho = np.sum(f * den)

    return cvti, chti, rho
```

```python
# Médias e parâmetros anisotrópicos para cada camada HTI
def media_camadasHTI(vp, vs, rho, depth, num_camadas, janela):

    limite = int((depth[-1] - depth[0])/janela)

    vp_dado, vs_dado, rho_dado, camadas = np.zeros(limite), np.zeros(limite), np.zeros(limite), np.zeros(limite)
    vp0, vs0, epsilon, gama, delta = np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas), np.zeros(num_camadas)

    i = 0
    for beds in range(num_camadas):
        for item in range(limite):
            if i < len(depth):
                vp_dado[item]  = vp[i]
                vs_dado[item]  = vs[i]
                rho_dado[item] = rho[i]
                camadas[item]  = depth[i]
                i = i + 1

        cvti, chti, rhobkus = bkusc(camadas[1:-2], vp_dado[1:-2], vs_dado[1:-2], rho_dado[1:-2])
        vp0[beds], vs0[beds], epsilon[beds], gama[beds], delta[beds] = parametros(chti, rhobkus)


    return vp0, vs0, epsilon, gama, delta
```

```python
# Abordagem Li (2006)

def parametros_anisotropicos(vsh, vp, vs):
    epsilon_li, gama_li, delta_li = np.zeros(len(vsh)), np.zeros(len(vsh)), np.zeros(len(vsh))
    for i in range(len(vsh)):
        epsilon_li[i] = 0.6*vsh[i]*(vp[i] - 1.5)/(6.05 - 1.5 - 2.65*vsh[i])
        gama_li[i]    = 0.67*vsh[i]*vs[i]/(4.09-2.29*vsh[i])
        delta_li[i]   = 0.32*epsilon_li[i]

    return epsilon_li, gama_li, delta_li
```

```python
# Parâmetros de Thonsem em função do volume da fração de vsh

def thomsen_from_vsh(vsh, vp, vs):
    epsilon_th, gama_th, delta_th = np.zeros(len(vsh)), np.zeros(len(vsh)), np.zeros(len(vsh))
    for i in range(len(vsh)):
        epsilon_th[i] = 0.23*vsh[i]
        gama_th[i]    = 0.38*vsh[i]
        delta_th[i]   = 0.1*vsh[i]

    return epsilon_th, gama_th, delta_th
```
