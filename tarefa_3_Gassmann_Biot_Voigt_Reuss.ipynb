#Cálculo do módulo de incompressibilidade
$$K_{s a t}=K_{d r y}+\frac{\left(1-\frac{K_{d r y}}{K_{\min }}\right)^2}{\frac{\phi}{K_{f l}}+\frac{(1-\phi)}{K_{\mathrm{min}}}+\frac{K_{d r y}}{K_{\mathrm{min}}^2}}$$

Módulos de incompressibilidade:

$K_{\text {sat }}=$ rocha saturada (poro não drenado)

$K_{d r y}=$ rocha seca (poro sem nenhum tipo de fluido)

$K_{\text {min }}=$ mineral da matriz

$K_{f l}=$ fluido (presente no poro)

$$K=\rho_{B}\left(v_{P}^{2}-\frac{4}{3} v_{S}^{2}\right) \text {, }$$

$$\mu=\rho_{B} v_{S}^{2} \text {, }$$

$$\mu_{s a t}=\mu_{d r y} $$

#Cálculo das velocidades da rocha saturada modelo de Gasmann

$$V_{\mathrm{P}}=\sqrt{\frac{k+\frac{4}{3} \mu}{\rho}}=\sqrt{\frac{\frac{k_{\mathrm{dry}}+\left(1-\frac{k_{\mathrm{dry}}}{k_{\mathrm{s}}}\right)^{2}}{k_{\mathrm{fl}}+\frac{1-\phi}{k_{\mathrm{s}}}-\frac{k_{\mathrm{dry}}}{k_{\mathrm{s}}^{2}}}+\frac{4}{3} \mu_{\mathrm{dry}}}{(1-\phi) \rho_{\mathrm{s}}+\phi \cdot \rho_{\mathrm{fl}}}}$$

\

$$V_{\mathrm{S}}=\sqrt{\frac{\mu}{\rho}}=\sqrt{\frac{\mu_{\mathrm{dry}}}{(1-\phi) \rho_{\mathrm{s}}+\phi \cdot \rho_{\mathrm{fl}}}}$$

```python
import numpy as np
import pandas as pd


from io import StringIO

# Define the table as a string
table_data = """
Sample,Density Dry,Vp (Radial) Dry,Vs(Radial) Dry,Porosity,Permeabilidade (mD),Density Sat,Vp (Radial) Sat,Vs (Radial) Sat
AB2,2214.162835,3721.618954,2365.119197,20.15,5.96,2422.546207,3850.868233,2376.8923
AB4,2128.969011,3459.633028,2291.008505,23.42,6.17,2364.802857,3657.613967,2303.78659
BC1,2239.489443,4068.965517,2463.144162,18.94,4.82,2424.505795,4154.015402,2495.3478
BC2,2209.519454,3831.129196,2404.853129,19.42,5.72,2414.237486,3908.260689,2420.6545
BC4,2120.34854,3468.26127,2472.131148,22.92,14.9,2358.81981,3584.672435,2600
BC8,2125.229544,3477.460902,2139.219015,23.07,5.64,2356.833365,3589.74359,2158.2
BC12,2271.428403,4370.37037,2539.340955,18.34,1.18,2447.95582,4416.374269,2554.803789
CD9,2064.000631,3314.360771,2070.568928,24.24,4.47,2317.481052,3428.442029,2104.45987
EF02,2002.204663,3282.722513,2109.927089,25.08,1.74,2255.019028,3476.89464,2200
"""

# Use StringIO to simulate reading from a file-like object
table_file = StringIO(table_data)

# Read the table into a Pandas DataFrame
df = pd.read_csv(table_file)

# Extraindo variáveis para vetores numpy

sample = df['Sample'].values
density_dry = df['Density Dry'].values
vp_dry = df['Vp (Radial) Dry'].values
vs_dry = df['Vs(Radial) Dry'].values
porosity = df['Porosity'].values
permeability = df['Permeabilidade (mD)'].values
density_sat = df['Density Sat'].values
vp_sat = df['Vp (Radial) Sat'].values
vs_sat = df['Vs (Radial) Sat'].values

# Função que define o cálculo de Gasmann


#   vp_dry = vp rocha seca
#   vs_dry = vs rocha seca
#   rho_dry = densidade bulk rocha seca
#   rho_fl_sat = densidade bulk do fluido na rocha saturada
#   K_fl_sat = módulo de incompressibilidade do fluido (presente no poro)
#   K_min = módulo de incompressibilidade do mineral
#   phi = porosidade

def gassman(vp_dry, vs_dry, rho_dry, rho_fl_sat, k_fl_sat, k_min, porosity):
    import numpy as np
    rho_sat = rho_dry + porosity*rho_fl_sat

    mi_dry = rho_dry*vs_dry**2
    k_dry = rho_dry*vp_dry**2-(4/3)*mi_dry

    k_sat = k_dry +( ((1-(k_dry/k_min))**2)/(porosity/k_fl_sat + (1-porosity)/k_min + k_dry/k_min**2))

    mi_sat = mi_dry

    vp_sat = np.sqrt(((k_sat + (4/3)*mi_sat)/rho_sat))
    vs_sat = np.sqrt((mi_sat/rho_sat))


    return vp_sat, vs_sat, rho_sat, k_sat, k_dry

#   vp_sat = vp rocha saturada
#   vs_sat = vs rocha saturada
#   rho_sat = densidade bulk rocha saturada
#   K_sat = modulo de incompressibilidade rocha saturada
#   K_dry = modulo de incompressibilidade rocha seca





```

```python
# ======================Constantes=============================================


rho_fl_sat = 997    # Densidade da água em kg/m³
rho_ank = 3100      # densidade Ankerita, varia de 2900 a 3100 kg/m³
rho_calc = 2710     # densidade Calcita em kg/m³

k_fl_sat = 2.05*1e9     # GPa modulo de incompressibilidade da água
k_ank = 70*1e9          # GPa módulo de incomp. ankerita varia entre 50 a 80 GPa
k_calc = 76*1e9         # GPa módulo de incomp. calcita

mi_ank = 30*1e9         # GPa módulo de cisalhamento ankerita varia de 20 a 30 GPa
mi_calc = 32*1e9        # GPa módulo de cisalhamento calcita

#============= Fração de mineral==============================================
frac_1 = (rho_min - rho_calc)/(rho_ank-rho_calc) #fração de ankerita
frac_2 = 1-frac_1                                #fração de calcita


k_min = frac_1*k_calc + frac_2*k_ank        # k_min = módulo de incompressibilidade do mineral em GPa
mi_min = frac_1*mi_calc + frac_2*mi_ank     # mi_min = módulo de cisalhamento do mineral

#==========================Vetores de densidade==================================

porosity = porosity/100                     # porosidade em decimal
rho_min = density_dry/(1-porosity)          # vetor de densidade seca do mineral em kg/m³
# density_dry = vetor de densidade seca das amostras experimentais em kg/m³
# density_sat = vetor de densidade de amostras saturadas experimentais em kg/m³
# rho_sat_gass = vetor densidade saturada por gassmann kg/m³

#=========================Vetores de Vp===========================================

# vp_dry = vetor VP das amostras experimentais secas em m/s
# vp_sat = vetor Vp das amostras experimentais saturadas em m/s
# vp_sat_gass = vetor VP das amostras por gassmann saturadas em m/s

#========================Vetores de Vs===========================================

# vs_dry = vetor VS das amostras experimentais secas em m/s
# vs_sat = vetor VS das amostras experimentais saturadas em m/s
# vs_sat_gass = vetor VS das amostras por gassmann saturadas em m/s

#=======================Vetores de Mod. de Incomp. ===================================

k_sat_exp = (vp_sat**2 - (4/3)*vs_sat**2)*density_sat
# k_dry = vetor modulo de incompressibilidade rocha seca experimental em GPa
# k_sat_exp = vetor modulo de incompressibilidade rocha saturada experimental em GPa
# k_sat_gass = vetor modulo de incompressibilidade rocha saturada por gassmann em GPa


```

#Cálculo modelo de Voigt-Reuss

$\begin{gathered}K_R=\left(\sum_{i=1}^N \frac{F_i}{K_i}\right)^{-1} \text { Reuss } \\ K_V=\sum_{i=1}^N F_i K_i \quad \text { Voigt }\end{gathered}$




```python

#===========================Calculo modelo Voigt-Reuss===========================

# Nesse modelo  acrescenta a fraçao do fluido que será a porosidade: 1 = Fração_ankerita + fração_calcita + Fração do fluido(porosidade)
f_ank = frac_1*(1-porosity)
f_cal = frac_2*(1-porosity)
#k_fl_sat = 2.05     # GPa modulo de incompressibilidade da água
#k_ank = 70          # GPa módulo de incomp. ankerita varia entre 50 a 80 GPa
#k_calc = 76         # GPa módulo de incomp. calcita

k_reuss = 1/((f_ank/k_ank) + (f_cal/k_calc) + (porosity/k_fl_sat))
k_voigt = f_ank*k_ank + f_cal*k_calc + porosity*k_fl_sat

#k_reuss = 1/((f_ank/k_ank*1e9) + (f_cal/k_calc*1e9) + (porosity/k_fl_sat*1e9))
#k_voigt = (f_ank/k_ank*1e9) + (f_cal/k_calc*1e9) + (porosity/k_fl_sat*1e9)


```

#Cálculo modelo Substituição de fluidos - Biot(Alta frequencia)

$V_{P \infty}=\left\{\frac{1}{\rho_0(1-\phi)+\phi \rho_{f l}\left(1-a^{-1}\right)}\left[\left(K_{f r}+\frac{4}{3} \mu_{f r}\right)+\frac{\phi \frac{\rho}{\rho_f} a^{-1}+\left(1-\frac{K_{f r}}{K_0}\right)\left(1-\frac{K_f}{K_0}-2 \phi a^{-1}\right)}{\left(1-\frac{K_{f f}}{K_0}-\phi\right) \frac{1}{K_0}+\frac{\phi}{K_{f l}}}\right]\right\}^{\frac{1}{2}}$

$V_{S \infty}=\left\{\frac{\mu_{f r}}{\rho_0(1-\phi)+\phi \rho_{f l}\left(1-a^{-1}\right)}\right\}^{\frac{1}{2}}$

\begin{array}{ll}
K_{f r}, \mu_{f r} \rightarrow \text { rocha seca } & \rho_0 \rightarrow \text { min eral } \\
K_0 \rightarrow \text { grão mineral } & \rho_f \rightarrow \text { fluido } \\
K_{f f} \rightarrow \text { fluido } & \rho \rightarrow \text { efetiva } \\
\phi \rightarrow \text { porosidade } & a^{-1} \rightarrow \text { tortuosidade }
\end{array}

```python

#===========================Calculo modelo Substituição de fluidos - Biot(Alta frequencia)===========================

def biot(Vp_dry, Vs_dry, rho_min, K_min, rho_fl, K_fl, phi, alpha):

    #The approximate high frequency limiting velocities predicted
    #Inputs: Vp_dry, Vs_dry: P and S velocities of dry rock
    #K0, MU0, RO0: Mineral bulk and shear moduli, and density
    #ROFL,KFL: Pore fluid density and bulk modulus
    #POR: Porosity
    #ALFA: Tortuosity parameter in Biot theory. Always > 1. Usually 1 to 3.

    rho_dry = (1 - phi)*rho_min
    rho_sat = (1 - phi)*rho_min + phi*rho_fl_sat
    mi_dry = rho_dry*Vs_dry**2
    K_dry = rho_dry*Vp_dry**2 - (4/3)*mi_dry
    b = K_dry/K_min
    rho_biot = rho_min*(1 - phi) + phi*rho_fl*(1 - 1/alpha)
    t1 = phi*rho_sat/(rho_fl*alpha) + (1 - b)*(1 - b - 2*phi/alpha)
    t2 = (1 - b - phi)/K_min + phi/K_fl
    Vp_sat = np.sqrt((1/rho_biot)*(K_dry + (4/3)*mi_dry + t1/t2))
    Vs_sat = np.sqrt(mi_dry/rho_biot)

    return rho_sat, Vp_sat, Vs_sat
alpha = 2
rho_biot, vp_sat_biot, vs_sat_biot = biot(vp_dry, vs_dry, rho_min, k_min, rho_fl_sat, k_fl_sat, porosity, alpha)
```
